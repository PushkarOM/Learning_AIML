#### What is Artificial Neural Network ?

It's a really rough or crude electronic model based on the neural structue of the brain.

#### Traning of ANNs.

Two ways -- supervised and unsupervised.


Supervised Traning
Here both the inputs and Output are provided for a set of data. The network then processes the inputs and compares it's output by the desired output. Errors are then Propogated back through the system, causing the system to adjust weights which control the netwroks output.

Converge -- a term used to indicate that the model provided desired output.

Sometime networks don't converge:-
either the inputs provided do not related with the output desired.
dataset might be small. -- train test split (ideally this should happend, meaning enough data should be present that it can be split in 2 sets).
If an algo can simply not solve a problem the the structure likly needs a redesgin.

There are many way to adjsut the weights, one of the most comman being Backwards Error propogation.


Unsupervised or Adaptive learning
Here network is provided with input but not the desired output. The system then itself must decied what features it will use to group the data. -- self organization.
