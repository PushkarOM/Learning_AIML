```python
import pandas as pd
```

#### Pandas
A Python-based data manipulation library that provides tools to extract, clean, and present data efficiently.
2 main ways pandas store data :-
- **Series :** A Series is a one-dimensional array-like object that can hold data of any type.
- **DataFrame :** A DataFrame is a two-dimensional, tabular data structure with labeled axes (rows and columns).

```python
# Creating a Series

pricelist = [100,200,300,400,500]

productseries = pd.Series(pricelist, index = ['pencil','rubber','earphones','mobile','laptop']) # if index list is not provided then by default 0 to n-1 indexing occurs
print(productseries)
```

```python
# Creating a DataFrame
productdf =  pd.DataFrame([[100,200,300,400,500],[10,20,30,40,50]], columns = ['pencil','rubber','earphones','mobile','laptop'])

# Printing the dataframe
print(productdf)

# Printing it's Shape (rows x columns)
print("Rows and Columns:",productdf.shape)

# Printing it's size
print("Size of DataFrame: ",productdf.size)

# Printing the Name of Columns
print("Column names: ",productdf.keys())

# Adding Rows and Columns
productdf2 = pd.DataFrame([[60,70,80,90,100],[25,26,27,28,29]], columns = ['pencil','rubber','earphones','mobile','laptop'])
print("Second DataFrame:\n",productdf2)

# Appending rows via another DataFrame
productdf3 = pd.concat([productdf, productdf2], ignore_index=True)  # provide a list of dataframes to join, ignore_index resets the indexing in the new dataframe
print("Appended DataFrame:\n",productdf3)


# Adding Column named 'watch'.
productdf3['watch'] = [20,70,80,60]
print("Appended DataFrame:\n",productdf3)
print("Size of DataFrame: ",productdf3.size)
print("Size of DataFrame: ",productdf3.shape)

# Deleteing a Column
productdf3 = productdf3.drop(columns=['watch'],axis = 0)
print("After Deletion DataFrame:\n",productdf3)

# Deleting a Row
productdf3 = productdf3.drop(index=[0])
print("After Deletion DataFrame:\n",productdf3)
```

```python
# Importing Data
# Function like read_csv(), read_html(), read_json(), read_sql() and many more, allow reading of all these file formate by Pandas Dataframe.
# We will be consdiering the "LiverPatient.csv", related to indian liver patient.

liverdf = pd.read_csv("./Data/Indian Liver Patient Dataset (ILPD).csv")
liverdf.columns = ['Age','Gender','TB','DB','Alkphos','Sgpt','Sgot','TP','ALB','AG','LiverPatient']
print("Rows and Columns of Liverdf:", liverdf.shape)
print("Size of Liverdf:", liverdf.size)

# Printing the column names
print("Keys in Liverdf:", liverdf.keys())
print("columns in Liverdf:", liverdf.columns)

liverdf # leaving it like this produces a automated tabel created by jypter notebook, you can do this aldo by print or dfname.describe 
```

```python
# Extracting Information (like dtype pf each value, count of each valye, null or not-null) From this Dataset
print("liverdf info:\n",liverdf.info())
print("Details of Liverdf:\n",liverdf.describe())

# describe is differnt from describe() as the former prints the entire dataset and the later prints different statitical measure about the dataset.
```

```python
# Reading the DataFrame
print("Printing the First 5 values:\n",liverdf.head()) # by default it prints 5
print("Printing the Last 5 values:\n", liverdf.tail()) # by default it prints 5
print("Printing the Last 20 values:\n", liverdf.tail(20)) # here it will print 20 values

# A particular Column can be specified as well
print("Printing the First 5 values of column 'Age' and 'TB':\n",liverdf[['Age','TB']].head())

# We can specifically print a column's values as well
print("Printing the values of Column Age:", liverdf['Age'].values)

# Counting the total number of values in a column
print("Printing the values of Column Age:", liverdf['Gender'].value_counts())

# Converting the Values of a column to list
print('The list of Age:\n',liverdf['Age'].tolist())

```

```python
# Statistical Functions that can be applied on DataFrame

print("Mean Age: ",liverdf['Age'].mean())
print("Median Age: ",liverdf['Age'].median())
print("Max Age: ",liverdf['Age'].max())
print("Min Age: ",liverdf['Age'].min())
print("Sum of all the Ages: ",liverdf['Age'].sum())
print("Product of all th Ages: ",liverdf['Age'].product())
print("3 Smallest Values:\n",liverdf['Age'].nsmallest().head(3))
print("3 largest Values:\n",liverdf['Age'].nlargest().head(3))
```

```python
# Sorting the DataFrame on the Basis of the Column
print("Descending Order Based on Age:\n",liverdf.sort_values(by='Age',ascending=False))
print("ascending Order Based on Age:\n",liverdf.sort_values(by='Age',ascending=True))
```

```python
# Data Extraction using Relational Operators

#Finding all the rows with Column Gender as 'Male'
male_data = liverdf[liverdf['Gender'] == 'Male']
print("Data For Males:\n",male_data)

#Finding all the records where Age is greater than 50
age_data = liverdf['Age'] > 50
print("Data For Males:\n",liverdf[age_data])

# Adding Conditional Logics, age > 50 and Male
age50_male = (liverdf['Age'] > 50) & (liverdf['Gender'] == 'Male')  #Here & is and operator, Similarly |  can used as Or operator
print("Age > 50 and Male:\n",liverdf[age50_male])

# We can also find min,max etc for the subset of datacreated
print("Age > 50 and Male max Value:\n",liverdf[age50_male].max())
```

```python
# iloc indexers , this specifically give the info about the row and column number specified.

# Printing the 3rd Row
print("The 3rd record:\n",liverdf.iloc[2]) #specifiying only the row index

# Printing the 3rd Row and 5th column
print("the 3rd Row and 5th column:",liverdf.iloc[2,4])  #[row,col] index

# Printing Multiple Specified Row ans Column
print("Multiple Records:\n",liverdf.iloc[[1,3],[5,6]]) #[[row_num_1,row_num_2,....],[col_num_1,col_num_2,....]] index

# Printing a Range of Rows
print("Range of records:\n",liverdf.iloc[7:9,2:5]) #from 2ed to 4th column, starting from 7 to 8th row
```

```python
# loc indexers, this gives info about the index value that is specified

print("The 3rd row:\n",liverdf.loc[3])
print("Range Of Rows:\n",liverdf.loc[1:5,])
print("Retriving Different Multiple Rows:\n",liverdf.loc[[12,23,576]])
print("Retriving Range of Row with selected range of Column:\n",liverdf.loc[5:8, 'Age':'TB'])

# Relational Operator with loc
print("Displaying rows where Column AGE has value 54:\n",liverdf.loc[liverdf['Age'] == 54,['Age','TB']])
print("Displaying rows where Column AGE has value greater than 54:\n",liverdf.loc[liverdf['Age'] > 54,['Age','TB']])
print("Displaying rows where Column AGE has value Smaller than 54:\n",liverdf.loc[liverdf['Age'] < 54,['Age','TB']])
# We can further combine the condition with logics
print("Displaying rows where Column AGE has value smaller than 54, and Gender is Male:\n", liverdf.loc[((liverdf['Age'] < 54) & (liverdf['Gender'] == 'Male')), ['Age', 'TB']])
```

```python
# Specific Function with loc indexer
print("Using Startswith Function:\n",liverdf.loc[liverdf['Gender'].str.startswith("Fe")&(liverdf['ALB']>=5)])
print("Using isin Function:\n",liverdf.loc[liverdf['ALB'].isin([4.4,4.2,4.3])&(liverdf['Age']>=60)])
```

```python
# Groupby functionality

# Grouping patients on the basis "LiverPatient" and finding the max values on the basis of DB.
print('Grouping on basis of "LiverPatient:\n"',liverdf['DB'].groupby([liverdf['LiverPatient']]).max())

print('Grouping on basis of "LiverPatient:\n"',liverdf['DB'].groupby([liverdf['LiverPatient']]).min())
```

```python
# Creating Charts Via Pandas

#Pie chart for "TB", labels of liver patient of 10 records.
data1 = liverdf.iloc[1:10,]
data1.groupby('TB').size().plot.pie(labels=data1['TB'], autopct='%1.1f%%')

# Scatter plot
liverdf.plot.scatter('TB','DB')

# Histogram Plot
liverdf.hist(column='Alkphos',by='Gender')

# Boxplot
liverdf.boxplot(column='AG',by = 'LiverPatient')

# CrossTab can be used for determing observation of 2 categorical values
print(pd.crosstab(liverdf.Gender,liverdf.LiverPatient,margins=True))

# plotting this CrossTab
pd.crosstab(liverdf.Gender,liverdf.LiverPatient,margins=True).plot(kind='bar',figsize=(7,5))
```

```python
# Finding Missing Values
print("Total Missing ValuesL\n",liverdf.isnull().sum())
```

```python
# 4 values are missing for the column 'AG'
# let's remove the missing values

liverdfcpy1 = liverdf.copy()
print("Original df: ",liverdfcpy1.loc[(liverdfcpy1['AG'].isnull()) == True])

# removing the Values
liverdfcpy1.dropna(inplace=True)

print("New df: ",liverdfcpy1.loc[(liverdfcpy1['AG'].isnull()) == True])
# We see an empty dataframe as the 'na' or 'null' values have been deleted.

liverdfcpy2 = liverdf.copy()

# Display rows where 'AG' is originally null
print("Original DataFrame where 'AG' is null:\n", liverdfcpy2.loc[liverdfcpy2['AG'].isnull()])

# Fill the null values in 'AG' with the median of 'AG'
liverdfcpy2['AG'] = liverdfcpy2['AG'].fillna(liverdfcpy2['AG'].median())  # This line effectivly return a SERIES, so i have explicilty assign it to a particular column.


# Check to confirm that no 'AG' values are null after filling
print("New DataFrame after filling 'AG' null values with median:\n", liverdfcpy2.loc[liverdf['AG'].isnull() == True])
```
