```python
import seaborn as sns
```

```python
# Using Seaborn Library to understand the "titanic dataset", categorical Data
titanic = sns.load_dataset("titanic")

print("Dimension of the Dataset: ",titanic.shape)
print("The columns in the Dataset are:\n", titanic.keys())
print("The Dataset's description:\n",titanic.describe())
```

```python
# Box plot
sns.boxplot(x = titanic['class'], y = titanic['fare'], hue="sex",data = titanic).set_title('Box Plot for fare with respect to class')
```

```python
# Violin Plot
sns.violinplot(x="class",y="age",hue="sex",data=titanic).set_title("Violin Plot for showing Age with class sex")
```

```python
# Point Plot
sns.pointplot(y = titanic['age'],x = 'class',data=titanic).set_title('Point Plot for age with respect to class')
```

```python
# Line Plot
sns.lineplot(y = titanic['fare'],x ='embarked',data=titanic).set_title('Line Plot for fare and embarked')
```

```python
# Count Plot
import matplotlib.pyplot as plt  # to show that seaborn is built to compliment matplotlib not to replace it.
sns.countplot(y = "alive", data = titanic)
plt.title("Count plot for alive")
```

```python
# Bar Plot
sns.barplot(y = 'fare', x = 'embarked', data=titanic).set_title('Bar Plot for fare with respect to embarked')
```

```python
# Strip Plot
sns.stripplot(x = "class", y = "fare", jitter = True, data = titanic).set_title("StripPlot for class vs fare")  # jitter = True to prevent overlapping points
```

```python
# SwarPLot
sns.swarmplot(x = "class", y = "fare", data = titanic).set_title("SwarmPlot for class vs fare")
```

```python
# catPlot
sns.catplot(x = "sex", y = "age", hue = "alive" , data = titanic)

# using the functin with "kind" argument
# Violin
sns.catplot(x = "sex", y = "age", hue = "alive" ,kind="violin" ,data = titanic)
# point
sns.catplot(x = "sex", y = "age", hue = "alive" ,kind="point" ,data = titanic)


# CatPlot with kind = "box", col = "class"
sns.catplot(x = "sex", y = "age", hue = "alive" ,kind = "box", col = "class",data = titanic)

# CatPlot with kind = "bar", col = "deck", col_wrap = 3
sns.catplot(x = "sex", y = "age", hue = "alive" ,kind = "bar", col = "deck", col_wrap = 3 ,data = titanic)
```

```python
import matplotlib.pyplot as plt
# Facet Grid

# Creating a FacetGrid
sns.FacetGrid(titanic,col = "class")

# Creating a Facet Grid for "age" and "class"
grid1 = sns.FacetGrid(titanic, col = "class")
grid1.map(plt.hist,"age")

# Creating a Facet Grid for "age", "class", "alive"
grid2 = sns.FacetGrid(titanic, col = "class", hue = "alive")
grid2.map(plt.hist,"age")

# Creating a Facet Grid for "age", "class", "alive"
grid3 = sns.FacetGrid(titanic, col = "class", hue = "alive")
grid3.map(plt.scatter,"age","fare")
```

```python
# Using the seaborn lib to understand the "iris dataset" , continious variables
iris = sns.load_dataset('iris')
print("Dimension of the Dataset: ",iris.shape)
print("The columns in the Dataset are:\n", iris.keys())
print("The Dataset's description:\n",iris.describe())
```

```python
# Scatter Plot
sns.scatterplot(y = "sepal_length", x = "petal_length", data = iris).set_title('Scatter Plot for sepal_legnth and petal_length')
```

```python
# Regression Plot

# regplot()
sns.regplot(y = "petal_width", x = "petal_length", data=iris)

# lmplot()
sns.lmplot(y = "sepal_width", x ="petal_width", data= iris)

# Higher Order visualization
sns.lmplot(y = "sepal_width", x = "petal_width", data = iris, order = 2)
```

```python
# Heat Map
sns.heatmap(iris[['petal_length','petal_width']]).set_title("Heat Map fro petal_length and petal_width")
plt.show()
# specifying the color
sns.heatmap(iris[['petal_length','petal_width']],cmap = "Blues").set_title("Heat Map fro petal_length and petal_width")
```

```python
# Univariate Distribution Plot, histplot() {formerly distplot()}
plt.figure(figsize = (10,10))

# Univariate distribution plot for "petal_length"
plt.subplot(3,1,1)
sns.histplot(iris[['petal_length']], kde = True).set_title('Distribution PLot for petal_legnth')

# without Kde
plt.subplot(3,1,2)
sns.histplot(iris[['petal_length']], kde = False).set_title('Distribution PLot for petal_length')

# displaying only the kde (kernal density estimate)
plt.subplot(3,1,3)
sns.kdeplot(iris[['petal_length']]).set_title('Distribution PLot for petal_length')
```

```python
# Joint Plot
sns.jointplot(y = "petal_length", x = "sepal_length", data = iris)


# regression joint plot
sns.jointplot(y = "petal_length", x = "sepal_length", data = iris, kind = 'reg')
```

```python
# Joint Hexbin Plot
sns.jointplot(y = "petal_length", x = "sepal_length", kind="hex", data = iris)
```

```python
# Joint kernel Density Plot
sns.jointplot(y = "petal_length", x = "sepal_length", kind="kde", data = iris)
```

```python
# Pair Plot

# Pair plot using a regression line
sns.pairplot(iris, vars = ['sepal_length','sepal_width','petal_length', 'petal_width'], kind = "reg")

#Pair Plot for different species-selected columns of dataset
sns.pairplot(iris, vars = ['sepal_length','sepal_width','petal_length', 'petal_width'],hue = "species", kind = "reg", diag_kind= 'hist', palette = 'colorblind')


# Pair plot using regression line for entire dataset
sns.pairplot(iris, kind = 'reg')

# Pair plot for different species and kde as diagonal plot
sns.pairplot(iris, kind = 'scatter',diag_kind= "kde", palette='deep',hue = 'species')
```

```python
# Pair Grid with scatter plot
grid1 = sns.PairGrid(iris)
grid1.map(plt.scatter)

# Pair Grid with kde,scatter,hist plot
grid2 = sns.PairGrid(iris)
grid2.map_upper(plt.scatter)
grid2.map_lower(sns.kdeplot)
grid2.map_diag(plt.hist)
```
